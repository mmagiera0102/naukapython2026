```python
testowanie=read.copy()
testowanie.info()
testowanie.describe()
```

```python
testowanie.isnull().sum()
tstowanieo=testowanie.dropna()
tstowanieo.isnull().sum()
```

```python
tstowanieo=tstowanieo[tstowanieo.Quantity>=0]
tstowanieo.info()
```

```python
#tstowanieo.CustomerID=tstowanieo.CustomerID.apply(lambda x: str(int(x)))
tstowanieo
tstowanieo[['Quantity','UnitPrice','CustomerID']]
tstowanieo[['Quantity','UnitPrice','CustomerID']].head(10)
tstowanieo[['Quantity','UnitPrice','CustomerID']][:10]
tstowanieo[tstowanieo['CustomerID']=='17850']
tstowanieo[tstowanieo.CustomerID=='17850']
tstowanieo[(tstowanieo['CustomerID']=='17850') & (tstowanieo['UnitPrice']>5)]
tstowanieo[(tstowanieo['CustomerID']=='17850') | (tstowanieo['Country']=='France')]
tstowanieo[tstowanieo['InvoiceNo'].notnull()]
```

```python
tstowanieo.groupby('CustomerID').count()
tstowanieo.groupby('CustomerID').size()
#tstowanieo['Revenue']=tstowanieo.Quantity*tstowanieo.UnitPrice
tstowanieo.groupby('CustomerID').aggregate({'Revenue':'mean','CustomerID':'size'})
```

```python
tstowanieo['dzien_tyg']=tstowanieo.InvoiceDate.dt.day
tstowanieo.head()
```

```python
tstowanieo.groupby('dzien_tyg')['Revenue'].sum()
wynik=tstowanieo.groupby('dzien_tyg').aggregate({'Revenue':'sum'})
```

```python
import seaborn as sns
sns.set()

wynik.plot(kind='bar')
```
