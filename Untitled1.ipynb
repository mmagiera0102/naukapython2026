```python
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt
import numpy as np
dane=pd.read_csv('sales_data_practice.csv')
dane['date']=pd.to_datetime(dane['date'])
```

```python
dane['month']=dane['date'].dt.month
dane.info()
Sprzedaż_miesiac=dane.groupby('month')['total_value'].sum()
```

```python
plt.plot(Sprzedaż_miesiac.index, Sprzedaż_miesiac.values)
plt.title('Sprzedaż miesiąc')
plt.xlabel('Miesiac')
plt.ylabel('Sprzedaz')
plt.show()
```

```python
product_sales = dane.groupby('product')['total_value'].sum()

plt.bar(product_sales.index, product_sales.values)

plt.title("Sprzedaż per produkt")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.hist(dane['total_value'], bins=20)

plt.title("Rozkład wartości zamówień")
plt.xlabel("Wartość zamówienia")
plt.ylabel("Liczba zamówień")
plt.show()
```

```python
pivot = dane.groupby(['month','product'])['total_value'].sum().reset_index()
pivot = pivot.pivot(index='month', columns='product', values='total_value')
pivot.plot()

plt.title("Sprzedaż miesięczna per produkt")
plt.show()
```
